# Step 3: Merging the datasets

The project originally consisted of multiple CSV files representing different components of a banking system. These datasets were merged using key identifiers such as, CustomerID, AccountID,TransactionTypeID, LoanID, BranchID to create a unified dataset for analysis.

## 3.1 Merge Customer Information

```python
customer_df = customers.merge(customer_types,on='CustomerTypeID',how='left')

customer_df = customer_df.merge(addresses,on='AddressID',how='left')
```

**The customers dataset was merged with the customer_types dataset using the CustomerTypeID to include information about the type of each customer. Then, the resulting dataset was merged with the addresses dataset using AddressID to add location details such as street, city, and country for each customer.**

## 3.2 Merge Account Information

```python
account_df = accounts.merge(account_types, on='AccountTypeID',how='left')

account_df = account_df.merge(account_status, on='AccountStatusID',how='left')
```

**The accounts dataset was merged with the account_types dataset using AccountTypeID to include information about the type of each account. Then, the resulting dataset was merged with the account_status dataset using AccountStatusID to add details about the current status of each account.**

## 3.3 Merge Transcations

```python
transactions_df = transactions.merge(transaction_types, on='TransactionTypeID', how='left')

transactions_df = transactions_df.merge(branches, on='BranchID', how='left')
```

**The transactions dataset was merged with the transaction_types dataset using TransactionTypeID to include the type of each transaction. Then, the resulting dataset was merged with the branches dataset using BranchID to add information about the branch where the transaction occurred.**

## 3.4 Merge Transactions with Accounts

```python
transactions_df = transactions_df.merge(account_df, left_on='AccountOriginID', right_on='AccountID',how='left')
```

**The transactions_df dataset was merged with the account_df dataset using AccountOriginID from the transactions table and AccountID from the accounts table. This merge adds account-related information such as account type, account status, and balance to each transaction record**

## 3.5 Merge Customers with Transactions


```python
master_df = transactions_df.merge(customer_df, on='CustomerID',how='left')
```

**The transactions_df dataset was merged with the customer_df dataset using CustomerID to include customer-related information for each transaction. This merge adds details such as the customer's name, demographic information, and address to the transaction records.**

## 3.6 Merge Loan Data

```python
loan_df = loans.merge(loan_status, on='LoanStatusID',how='left')

master_df = master_df.merge(loan_df, on='AccountID', how='left')
```

**The loans dataset was merged with the loan_status dataset using LoanStatusID to include information about the status of each loan. Then, the resulting loan_df was merged with the master_df using AccountID to add loan details such as loan amount, interest rate, and loan status to the master dataset.**

```python
df = master_df
df.to_csv("C:/Users/shubh/Desktop/Simran/Project/Prodigy-info-tech/Project-2/Finance_Fraud&Loan_Dataset/finance_Fraud_master_dataset.csv",index=False)
```

#### Step 7: Converting date columns

### Step 8: Data Cleaning & Preprocessing

#### 8.1 Renaming the columnns

```python
df = df.rename(columns={"TypeName_x":"TransactionType","TypeName_y":"AccountType","TypeName":"CustomerType","StatusName_x":"AccountStatus","StatusName_y":"LoanStatus"})
```

```python
df.columns
```

#### 8.2 Creating new features
- Loan Duration Days: Calculated by finding the difference between EstimatedEndDate and StartDate, representing the total duration of each loan in days.
- Transaction Year: Extracted from TransactionDate to analyze transaction patterns across different years.
- Transaction Month: Extracted from TransactionDate to study monthly transaction trends and seasonal patterns.
- Customer Age: Computed using the difference between the current date and DateOfBirth to analyze customer demographics and financial behavior across age groups.

```python
df['LoanDurationDays'] = (df['EstimatedEndDate'] - df['StartDate']).dt.days
df['TransactionYear'] = df['TransactionDate'].dt.year
df['TransactionMonth'] = df['TransactionDate'].dt.month
df['CustomerAge'] = (pd.Timestamp.now() - df['DateOfBirth']).dt.days // 365
```

```python
df.columns
```

## 8.3 Handling Missing Values in the dataset

#### 8.3.1. Loan related missing values
- Principal Amount: Missing values in the PrincipalAmount column were replaced with 0, as many transactions in the dataset are not linked to a loan, meaning there is no principal amount involved.

- Interest Rate: Missing values in the InterestRate column were filled with 0 to indicate that no interest rate applies when a loan is not present.

- Loan Duration Days: Any missing values in the LoanDurationDays column were set to 0, showing that the transaction or account does not have an associated loan duration.

- Loan Status: Missing values in the LoanStatus column were replaced with "No Loan" to clearly reflect that the customer or account does not currently have a loan.


```python
df['PrincipalAmount'] = df['PrincipalAmount'].fillna(0)
df['InterestRate'] = df['InterestRate'].fillna(0)
df['LoanDurationDays'] = df['LoanDurationDays'].fillna(0)
df['LoanStatus'] = df['LoanStatus'].fillna("No Loan")
```

#### 8.3.2 Handling Missing Address Information

- Missing values in the Street, City, and Country columns were replaced with "Unknown" to indicate that the address details were not available in the dataset.

- This approach helps retain all records in the dataset while clearly marking where location information is missing.

```python
address_cols = ['Street','City','Country']

for col in address_cols:
    df[col] = df[col].fillna("Unknown")
```

#### 8.3.3 Handling Missing Customer Names
- Missing values in the FirstName and LastName columns were replaced with "Unknown" to indicate that the customer's name information was not available in the dataset.

- This ensures that all records remain in the dataset while clearly marking where customer name details are missing.

```python
df['FirstName'] = df['FirstName'].fillna("Unknown")
df['LastName'] = df['LastName'].fillna("Unknown")
```

#### 8.3.4 Handling Missing Transaction Dates

- Records with missing values in the TransactionDate column were removed from the dataset.
- Since the transaction date is essential for time-based analysis, rows without this information were dropped to ensure accurate temporal analysis.

```python
df = df.dropna(subset=['TransactionDate'])
```

#### 8.3.5  Handling Opening Date Missing Values

- Missing values in the OpeningDate column were filled using the median opening date from the dataset.
- This helps keep the dataset complete while maintaining a realistic representation of when most accounts were opened.

```python
df['OpeningDate'] = df['OpeningDate'].fillna(df['OpeningDate'].median())
```

#### 8.3.4 Handling Missing Customer Age
- Missing values in the CustomerAge column were replaced with the median age of customers in the dataset.
- Using the median helps keep the data realistic while preventing missing values from affecting further analysis.


```python
df['CustomerAge']=df['CustomerAge'].fillna(df['CustomerAge'].median())
```
